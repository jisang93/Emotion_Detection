## 모듈 불러오기

```python
# 모듈 불러오기
from urllib.request import urlretrieve
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from selenium import webdriver
import os
```

## 검색어 입력

```python
# 검색어(ex. 웃는 얼굴, 행복한 표정 등) 입력하여 url 작성.
search = input('검색어를 입력해 주세요')
search_re = quote_plus(search)
url = 'https://www.google.com/search?q={}&sxsrf=ALeKk01352q8Q5113NFpa8AnaucxFnvwnA:1594094481772&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjH7cvaoLrqAhXPQN4KHfiEDvQQ_AUoAXoECAsQAw&biw=1536&bih=722'.format(search_re)

# 위 url로 접속이 안되시면 url을 수정해주셔야 합니다.
# 아무 검색어로 직접 구글에서 이미지 검색하시고, 
# 해당 url에서 검색어 부분에 위의 search_re 변수를 넣는 format 문자열로 url을 수정해주세요.
```

## 크롬 드라이버로 이미지 url 가져오기

```python
# 크롬 드라이버 열어서 url 열기
chromedriver = 'C:/webdriver/chromedriver' # chromedriver 파일 있는 위치
driver = webdriver.Chrome(chromedriver)
driver.get(url)

# 스크롤 내리기 - 스크롤 한번마다 400장* 5번 = 2000장까지 검색. 
# 그 이상은 검색이 되지 않는 경우도 많고 나와도 
# 검색어와 연관성 많이 떨어지는 이미지들이 나와서 제한.
for n in range(5):
    # 스크롤 최대한 내리기
    for i in range(500):
        driver.execute_script("window.scrollBy(0, 100000)")
        
    # 결과 더보기 클릭(검색된 이미지 수 2000장 보다 적으면 에러 나와서 예외 처리).
    try:
        element = driver.find_element_by_css_selector('#islmp > div > div > div > div > div.YstHxe > input')
        element.click()
        driver.implicitly_wait(5)
    except:
        pass
```

```python
# 이미지 주소 가져오기
html = driver.page_source
soup = BeautifulSoup(html)
img = soup.select('img.rg_i.Q4LuWd')
imgurl = []

for i in img:
    try:
        imgurl.append(i.attrs['src'])
    except:
        imgurl.append(i.attrs['data-src'])
len(imgurl)
```

## 이미지 저장

```python
# 검색어 이름으로 폴더 생성 후에 폴더 내에 이미지 저장 
os.mkdir('{}'.format(search))
n = 1
for i in imgurl:
    urlretrieve(i, "{0}/{0}_{1}.jpg".format(search, str(n)))
    n+= 1
    if n%100 == 0:
        print('{}장 저장 완료'.format(n))
    
driver.close()
print('{}장 저장 완료'.format(n)
```
